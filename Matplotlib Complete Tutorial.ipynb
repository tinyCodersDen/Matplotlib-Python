```python
# Plotting a point in matplotlib:
import matplotlib.pyplot as plt
plt.scatter([1,2,5],[3,4,2]) # Plots points at (1,3),(2,4),(5,2)
plt.show()
```

```python
# Plotting a line in matplotlib:
plt.plot([3,8,1,10])
plt.show()
```

```python
# Colors of lines:
plt.plot([3,8,1,10], color = 'r')
plt.show()
```

```python
# You can even have hexadecimal color values!:
plt.plot([3,8,1,10], c = '#4CAF50')
plt.show()
```

```python
# Setting the line width:
plt.plot([3,8,1,10], linewidth = '20.5')
plt.show()
```

```python
# Having multiple lines:
plt.plot([3,8,1,10])
plt.plot([5,2,8,11])

plt.show()
```

```python
# Having legends:
plt.plot([3,8,1,10],label='Line1')
plt.plot([5,2,8,11],label='Line2')

plt.legend()
plt.show()
```

```python
# Markers in matplotlib:
plt.plot([3,8,1,10],marker='o')
plt.show()
```

```python
# Labeled axis and titles:
plt.xlabel("X-Axis label here")
plt.ylabel("Y-Axis label here")
plt.title('Title here')
plt.show()
```

```python
# Gridlines in matplotlib:
plt.plot([1,2],[3,4])

plt.grid()

plt.show()
```

```python
# Subplots in matplotlib:
plt.subplot(1, 2, 1)
plt.plot([1,2,5],[3,4,7])
plt.subplot(1, 2, 2)
plt.plot([5,6,7],[5,1,6])
```

```python
# Bar graphs in matplotlib
plt.bar(['A','B','C'],[3,5,1])
plt.show()
```

```python
# Histograms in matplotlib:
import numpy as np
x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

```python
# Pie charts in matplotlib:
plt.pie([35, 25, 25, 15])
plt.show()
```
