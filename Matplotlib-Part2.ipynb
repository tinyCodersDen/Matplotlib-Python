```python
# Gridlines in matplotlib:
import matplotlib.pyplot as plt
plt.plot([1,2],[3,4])

plt.grid()

plt.show()
```

```python
# Subplots in matplotlib:
plt.subplot(1, 2, 1)
plt.plot([1,2,5],[3,4,7])
plt.subplot(1, 2, 2)
plt.plot([5,6,7],[5,1,6])
```

```python
# Bar graphs in matplotlib
plt.bar(['A','B','C'],[3,5,1])
plt.show()
```

```python
# Histograms in matplotlib:
import numpy as np
x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

```python
# Pie charts in matplotlib:
plt.pie([35, 25, 25, 15])
plt.show()
```
