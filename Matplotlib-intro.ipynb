```python
# Plotting points in matplotlib
import matplotlib.pyplot as plt
plt.scatter([1,8,6,5],[4,5,9,11])
plt.show()
```

```python
# Markers in matplotlib:
plt.scatter([1,8,6,5],[4,5,9,11], marker='^')
plt.show()
# As you see in the plot below, I am plotting triangles instead of circles.
# For a list of all matplotlib markers(including the one below),see here: https://matplotlib.org/stable/api/markers_api.html
```

```python
# Plotting lines in matplotlib:
plt.plot([1,8,9,14])
plt.show()
```

```python
# Lines with markers:
plt.plot([1,8,9,14],marker='o')
plt.show()
```

```python
# Labeling your axis in plots:
plt.plot([1,2],[3,4])

plt.xlabel("X-axis here")
plt.ylabel("Y-axis here")

plt.show()
```

```python
# Setting a title for your plot
plt.title("Title here")
plt.plot([1,2],[3,4])
plt.show()
```
